```python
# 3. Schroeder problem 1.7

n, x = symbols ('n x')
A = 1
a = 1 

psi = A*sin(n*pi*x/a)

p1 = plot(psi.subs(n,1/2), psi.subs(n,1), psi.subs(n,3/2), (x,0,2), legend=True, show=False)

p1[0].line_color='black'
p1[1].line_color='purple'
p1[2].line_color='orange'

p1[0].label='1st'
p1[1].label='2ed'
p1[2].label='3rd'

p1.title = ('eigenfunctions of infinite square well @ HSU')

p1.show()
```

```python
# 4. Schroeder 1.5

print('All wavelengths are in meters!\n')

hc  = 1240e-9 # eV*m*C
h = 6.63e-34 # J*s

# a) microwave photon
E_mw = 10e-5 # eV
lamb_mw = hc/E_mw
print('a) microwave photon\n', lamb_mw, 'm\n')

# b) electron
PC_e = 5000*1.60217663e-19 *6.242e+18 # 5kV to eV
lamb_e = hc/PC_e
print('b) electron\n',lamb_e,'m\n')


# c) rubidium-87 atom
KE_87 = 10e-11 # eV
PC_87 = sqrt(2*87*KE_87)  # non-relativistic p=sqrt(2m*KE)? What's the correct mass?
lamb_87 = hc/PC_87
print('c) rubidium-87 atom\n', lamb_87, 'm the # looks off :(\n')

# d) proton
PC_p = 6.5e12
lamb_p = hc/PC_p
print('d) proton\n', lamb_p, 'm\n')

# e) gnat
m = 0.2e-3 # grams
v = 0.5 # m/s
PC_KE = 0.5*m*v*v
lamb_gnat = h/(PC_KE*3e8)
print('e) gnat\n', lamb_gnat, 'm\n')
```
